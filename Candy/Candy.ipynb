```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_excel(r'C:\Users\omar mohamed\Downloads\task\candyhierarchy2017.xlsx')
```

```python
#heatmap to show null value
plt.figure(figsize=(15,12))
sns.heatmap(df.isnull(),yticklabels=False)
#too many null :D
```

```python
#check for duplicated
df.duplicated().sum()
```

```python
#check for null
df.isnull().sum()
```

```python
#specify columns which have null more than 900
null = df.isnull().sum()[df.isnull().sum() >=900]
null
```

```python
#delete columns which have 900 or more null
for col in null.index:
    df.drop(col,axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
#edit col name
x = []
for i in  df.columns:
    if i.startswith('Q'):
        i=i.replace(i[0:3],'').strip().title()
        if i.startswith('|') or i.startswith(':'):
            i=i.replace(i[0],'').strip()
           
        
    else :
        i = i.title()
    x.append(i)
df.columns = x
    
```

```python
df.tail(5)
```

```python
#fill null with mode
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['State, Province, County, Etc'].fillna(df['State, Province, County, Etc'].mode()[0],inplace=True)
df['Going Out?'].fillna(df['Going Out?'].mode()[0],inplace=True)
df['Country'].fillna(df['Country'].mode()[0],inplace=True)
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
```

```python
#see the result
df.isnull().sum()
```

```python
#change data to category type
df['Gender']=df['Gender'].astype('category')
df['Going Out?']=df['Going Out?'].astype('category')
```

```python
df.info()
```

```python
#fill rest of data by mode
for i in df.columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.head(10)
```

```python
#edit names in country col
df['Country'] = df['Country'].str.lower().str.strip()
```

```python
#change all usa names to united states
df['Country'][df['Country'].str.startswith('us') | df['Country'].str.startswith('u.s') ] = 'United States'
```

```python
df['Country'].fillna(df['Country'].mode()[0],inplace=True)
```

```python
#see the result of fill the null
plt.figure(figsize=(10,8))
sns.heatmap(df.isnull(),yticklabels=False)
plt.show()
```

```python
#melt the data
df = df.melt(id_vars=['Internal Id','Going Out?','Gender','Age','Country','State, Province, County, Etc','Dress','Day','Click Coordinates (X, Y)'],var_name='Candy Kind',value_name='Kind')
```
