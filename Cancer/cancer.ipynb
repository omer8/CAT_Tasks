## Delete all outliers

```python
for col in df.columns:
    df1=df[[col]]
    Q1 = df1.quantile(0.25)
    Q3 = df1.quantile(0.75)
    IQR = Q3-Q1
    minimum = Q1 - 1.5*IQR
    maximum = Q3 + 1.5*IQR
    condition = (df1 <= maximum) & (df1 >= minimum)
    condition = condition.all(axis=1)
    df=df[condition]
    sns.distplot(df[col],kde=False)
    plt.show()
```
