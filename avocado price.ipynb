```python
#pairplot to explor the data
sns.pairplot(df)
```

```python
#groub the data by the year
df1 = df.groupby('year')
```

```python
#compute sum for volume in each year and mean for price in each year
df1_year=df1.agg({'Total Volume':'sum','AveragePrice':'mean'})
df1_year
```

```python
#plot average price and total volume in each year
plt.figure(figsize=(10,8))
ax1 = plt.subplot(1,2,1)
plt.plot(df1_year.index, df1_year['AveragePrice'],marker='s')
plt.xticks(df1_year.index)
ax1.set_xticklabels(['2015','2016','2017','2018'])
plt.xlabel("Year")
plt.ylabel("average price")
plt.title("Average Price per Year")
ax2 = plt.subplot(1,2,2)
plt.plot(df1_year.index, df1_year['Total Volume'],marker='o',color='red')
plt.xticks(df1_year.index)
ax2.set_xticklabels(['2015','2016','2017','2018'])
plt.xlabel("Year")
plt.ylabel("total volume")
plt.title("Total Volume per Year")
```
