```python
df1 = df.groupby('species')
```

```python
df1_mean = df1.mean()
df1_mean
```

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.plot(df1_mean.index,df1_mean['sepal_length'],marker='s')
plt.ylabel('sepal_length')
plt.xlabel('species')
plt.title('Avg sepal_length for each Species')

plt.subplot(2,2,2)
plt.plot(df1_mean.index,df1_mean['sepal_width'],marker='o',color='orange')
plt.ylabel('sepal_width')
plt.xlabel('species')
plt.title('Avg sepal_width for each Species')

plt.subplot(2,2,3)
plt.plot(df1_mean.index,df1_mean['petal_length'],marker='*',color='green')
plt.ylabel('petal_length')
plt.xlabel('species')
plt.title('Avg petal_length for each Species')

plt.subplot(2,2,4)
plt.plot(df1_mean.index,df1_mean['petal_width'],marker='x',color='gray')
plt.ylabel('petal_width')
plt.xlabel('species')
plt.title('Avg petal_width for each Species')

plt.show()
```
